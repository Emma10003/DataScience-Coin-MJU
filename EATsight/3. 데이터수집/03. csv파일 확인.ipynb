```python
import pandas as pd
```

```python
data = pd.read_csv("romand_reviews.csv")
data.head(3)
```

```python
user_tags_list = data['user_tags'].str.split(',')
user_tags_list
```

- `strip(',')`: 앞뒤로 불필요한 쉼표가 있을 경우 제거 (ex. ",복합성,여름쿨톤," -> "복합성, 여름쿨톤")
- `split(',')`: 쉼표를 기준으로 분리해서 리스트를 만듦.
- strip()과 split()은  Pandas Series에 사용할 수 없는 속성이므로 앞에 .str를 붙여줌

Python의 문자열 객체에는 .split()이나 .strip() 같은 메서드가 기본으로 있음.  
그런데 Pandas의 Series는 여러 개의 문자열을 담은 일종의 열(column)이기 때문에, .split() 같은 문자열 메서드를 직접 호출할 수 없다!

그래서 Pandas는 .str 접근자를 제공  
-> Series.str은 Series 안의 **각 문자열 요소에 대해** 문자열 함수를 적용할 수 있게 해주는 벡터화된(string vectorized) 접근 방법.



```python
data['user_tags_list'] = user_tags_list
data.head(1)
```

```python
data = data.dropna(subset=['user_tags'])
data[data['user_tags_list'].apply(len) > 4]
```

- `apply()`: Pandas에서 Series나 DataFrame의 각 원소에 대해 특정 함수를 "하나씩" 적용할 때 사용하는 함수.
- `len()`은 리스트에는 적용할 수 있지만 Series 전체에는 쓸 수 없음. -> 쓸려면 Series 행 만큼 반복문을 돌려야 한다..
- `subset`: 어떤 컬럼을 기준으로 결측값을 판단해서 제거할지를 지정.
- DataFrame 내에서 태그가 4개보다 많은 행이 존재하는지 알아보기 위해서 'user_tags_list'행을 추가,  
`data['user_tags_list']`는 이름표만 붙여놓은 리스트 모음 -> 비교문을 통해 해당 코드는 불리언 Series가 됨,  
`data[...]`에 해당 불리언 Series를 넣어 값이 True인 행을 원본 데이터에서 필터링해 가져옴.
