```python
import pandas as pd
import numpy as np
```

```python
file_name = 'ys_crawling_data/forencos_soakTatto_1-2_reviews.csv'
df = pd.read_csv(file_name)
df.head(5)
```

```python
# 필요한 톤 키워드 리스트 정의
tone_keywords = ['봄웜톤', '여름쿨톤', '가을웜톤', '겨울쿨톤']

# 예: NaN 처리와 톤 추출을 한 줄로 처리
def extract_skin_tone(tags):
    if pd.isna(tags):
        return np.nan
    # 리스트 형태로 나눠서 톤 키워드만 필터링
    tones = [tone.strip() for tone in tags.split(',') if tone.strip() in tone_keywords]
    return ','.join(tones) if tones else np.nan  # 없으면 NaN 반환

# 새 컬럼 추가
df['skin_tone'] = df['user_tags'].apply(extract_skin_tone)

# 추가된 컬럼 확인
df[['user_tags', 'skin_tone']].head(5)
```

```python
# 원본 데이터에 새로운 컬럼이 제대로 추가되었는지 확인
df.head(1)
```

```python
# 같은 이름으로 다시 저장
df.to_csv(file_name, index=False, encoding='utf-8-sig')
```
