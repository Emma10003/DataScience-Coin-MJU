#### 데이터 개수

```python
# 전체 데이터 행 개수
print(f"전체 데이터 개수: {df.shape[0]:,}개")

# skin_tone 컬럼의 값이 결측치인 행 개수
skin_tone_NaN = df['skin_tone'].isna().sum()
print(f"피부톤 결측치 데이터 개수: {skin_tone_NaN:,}개")

# 정상 데이터 개수
normal_data = df.shape[0] - skin_tone_NaN
print(f"정상 데이터 개수: {normal_data:,}개")
```
